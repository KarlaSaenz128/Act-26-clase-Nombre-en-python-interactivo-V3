#### crear una publicacion , con 3 atributos(tipo,valor,unidad),2 funciones leerdatos(),mostrardatos() y 2 objetos(publicacion1,publicacion3).

#### no usar while true,if,for,listas y diccionario

```python
class Publicacion:
    def __init__(self, titulo, contenido, autor, fecha):
        # Inicializamos los atributos vacíos
        self.titulo = ""
        self.contenido = ""
        self.autor = ""
        self.fecha = ""

    def leerdatos(self):
        # Captura de los 4 datos por teclado
        print("\n--- Ingresando Información ---")
        self.titulo = input("Ingrese el título de la publicación: ")
        self.contenido = input("Ingrese el contenido: ")
        self.autor = input("Ingrese el autor: ")
        self.fecha = input("Ingrese la fecha (dd/mm/aaaa): ")

    def mostrardatos(self):
        # Exhibición de los 4 atributos
        print("Título:", self.titulo)
        print("Contenido:", self.contenido)
        print("Autor:", self.autor)
        print("Fecha:", self.fecha)

# --- Creación de Objetos (post1 y post2) ---

# Operaciones para el Post 1
print("--- DATOS DEL POST 1 ---")
post1 = Publicacion("", "", "", "")
post1.leerdatos()
print("-*-*-*-*")
post1.mostrardatos()
print("-*-*-*-*")

# Operaciones para el Post 2
print("\n--- DATOS DEL POST 2 ---")
post2 = Publicacion("", "", "", "")
post2.leerdatos()
print("-*-*-*-*")
post2.mostrardatos()
print("-*-*-*-*")

print("\nELABORADO por karla saenz0133")
```
